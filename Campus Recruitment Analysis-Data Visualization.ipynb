### 3.2 Data Visualisation

```python
# Boxplot of distribution of gender of students's as per their salaries

sns.boxplot('salary','gender', data=placement)
```

We see that the median salary of male students is higher than female students

```python
# Boxplot of distribution of work experience status of students over their salaries

sns.boxplot('salary','workex', data=placement)
```

We see that the median salary of experienced students, is, not surprisingly, higher than unexperienced students, there are some outliers in case of experienced students as some of them must have been offered even higher packages due to other extraordinary factors.

```python
# Boxplot of distribution of high school streams of students over their salaries

sns.boxplot('salary','hsc_s', data=placement)
```

We see that the median salaries of all categories is somewhat similar, however the max of science and commerce students is higher, but since we do not have enough data for arts students, it'll be difficult to comment on this aspect.

```python
# Boxplot of distribution of degree majors of students over their salaries

sns.boxplot('salary','degree_t', data=placement)
```

```python
# Boxplot of distribution of degree majors of students over their salaries

sns.boxplot('salary','specialisation', data=placement)
```

We see that the median salary of marketing & finance students is higher than median salaries of marketing & HR, even the minimum salary for finance people is similar to median salary for HR people. It clearly shows that Marketing & finance specialisation is preferred over the other by the recruiters.
